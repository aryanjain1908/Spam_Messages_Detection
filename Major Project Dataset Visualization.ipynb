```python
# importing required modules 

# this module helps us to combine two different classes and make pipeline
from sklearn.pipeline import Pipeline

# this module helps us to read text
from sklearn.feature_extraction.text import CountVectorizer

# this module helps us to find the accuracy score
from sklearn.metrics import accuracy_score

# importing decision tree classifier algorithm
from sklearn.tree import DecisionTreeClassifier

# this module helps us to split the dataset to the training and testing
from sklearn.model_selection import train_test_split

# importing multinomial ML algorithm
from sklearn.naive_bayes import MultinomialNB

# this module helps us to visulise the data
import matplotlib.pyplot as plt

# importing some commmon modules
import pandas as pd
import numpy as np
import sys
import re


# this function prints the data head and tails
def Show_Dataset_Head_And_Tails(data):
    
#     printing first 10 rows of our data
    print("---------------Printing First 10 Rows--------------------")
    print(data.head(10))
    
#     printing last 10 rows of our data
    print("---------------Printing Last 10 Rows--------------------")
    print(data.tail(10))
    
    print()
    print("Rows in our data : ",data.shape[0])
    print("Columns in our data : ",data.shape[1])
    print("Shape of data is : ",data.shape)


    
    
    
    
# this function cisualise the current data in a form of bar graph and pie graph that how much spam and ham messages are there
def Visualise_Current_Dataset(data):
    
#     counting spam messages from the data
    spam_count = data[data["Types"] == 'spam']["Types"].count()
    
#     counting ham messages from the data
    ham_count = data[data["Types"] == 'ham']["Types"].count()
    
#     making list of count
    count = [spam_count , ham_count]
    
#     making list of colors
    color = ['red' , 'green']
    
#     making list of names
    name = ["Spam" , "Ham"]
    
#     adh=justing subplots
    plt.subplot(1,2,1)
    
#     plotting horizontal lines for analysing
    plt.axhline(spam_count , 0 , 1 , linestyle='dashed' , color = 'red')
    plt.axhline(ham_count , 0 , 1 , linestyle = 'dashed' , color = 'green')
    
#     plotting bar graph 
    for i in range(2):
        plt.bar(name[i] , count[i] , color = color[i] , width = 0.2)
        
#     setting title for graph
    plt.title("Bar Graph Of Messages Count").set_fontsize(20)
    
#     setting x and y labels of graph
    plt.xlabel("Types of Messages").set_fontsize(15)
    plt.ylabel("Number of Messages").set_fontsize(15)
    plt.legend(count , bbox_to_anchor = (1.05,1))
    
#     show the bar graph
    plt.show()
    
#     adjusting subplot 2
    plt.subplot(1,2,2)
    
#     plotting pie graph
    plt.pie(count , labels = name , explode = [0,0.2] , autopct = "%1.1f%%" , shadow = True)
    
#     setting title for graph
    plt.title("Pie Chart Of Messages Count").set_fontsize(20)
    
    plt.legend(bbox_to_anchor = (1.05,1))
    plt.show()
    
    
    
    
    

    
    
# this function checking naive bayes gaussian algorithm
def Multinomial_Algorithm_Without_InBuilt_functions(data):
    
#     reading data otherwise this will change the data
    data = pd.read_csv('spam.csv' , names = ['Types' , 'Messages'])
    
#     using for loop i split messages into list
    for i in range(data['Messages'].count()):
        data.iloc[i,1] = data.iloc[i,1].split()

#     creating empty list 
    spam_list = []
    ham_list = []

#     iterating over data and according to message category extend this message words to list
    for i in range(data['Types'].count()):
        
        if data.iloc[i,0]=='ham':
            ham_list.extend(data.iloc[i,1])
            
        else:
            spam_list.extend(data.iloc[i,1])


#     counting how much spam and ham messages are there
    ham_count = np.array(data['Types']).tolist().count("ham")
    spam_count = np.array(data['Types']).tolist().count("spam")

#     calculating probability of ham and spam messages
    p_ham = ham_count / (ham_count + spam_count)
    p_spam = spam_count / (ham_count + spam_count)

#     creating empty dictionary
    ham = {}
    spam = {}

#     calculating probability of each word and store it
    for i in spam_list:
        
        if i not in spam:
            spam[i] = spam_list.count(i) / len(spam_list)

    for i in ham_list:
        
        if i not in ham:
            ham[i] = ham_list.count(i) / len(ham_list)

#     taking message as input from user which he wants to check
    mes = input("Enter Message which you want to check : ")
    mes = re.sub(r'[^\w\s]','',mes).lower().split()

#     initialising probability of meassage with respect to ham or spam and using for loop update it
    p_mes_ham = 1
    p_mes_spam = 1

    for i in mes:
        
#         if word of message not in dictionary then increase the count by 1 and calculate probability
        if i not in ham and i not in spam:
            p_mes_ham *= 1 / (len(ham_list) + 1)
            p_mes_spam *= 1 / (len(spam_list) + 1)
            
        elif i not in ham:
            p_mes_ham *= 1 / (len(ham_list) + 1)
            p_mes_spam *= spam[i]
            
        elif i not in spam:
            p_mes_ham *= ham[i]
            p_mes_spam *= 1 / (len(spam_list) + 1)
            
        else:
            p_mes_ham *= ham[i]
            p_mes_spam *= spam[i]

#     calculating final probability
    p_ham_mes = p_ham * p_mes_ham 
    p_spam_mes = p_spam * p_mes_spam 

#     normalising the values
    is_ham = p_ham_mes / (p_ham_mes + p_spam_mes)
    is_spam = p_spam_mes / (p_ham_mes + p_spam_mes)

#     checking whether the meaasge is spam or ham
    if is_ham > is_spam:
        print('ham')
        
    else:
        print("spam")

        
    
    
    
    
    

        
def Visualise_Dataset_Using_DecisonTreeClassifier_Algorithm(data):
    
#     creating x and y variable of our data set column 1 contains messages and column 0 contains types of messages
    x = data["Messages"]
    y = data["Types"]
    
#     making list of different classifier to make a pipeline
    algo = [('vect' , CountVectorizer()) , ('model' , DecisionTreeClassifier())]
    
#     making pipeline
    pipe = Pipeline(algo)
    
#     spliting data and giving test size is 1/3
    x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 1/3 , random_state = 0)
    
#     fitting x_tain and y_train in created pipe
    trained = pipe.fit(x_train , y_train)
    
#     predicting the output
    y_pred = trained.predict(x_test)
    
#     converting array to list
    y_pred = np.array(y_pred).tolist()
    y_test = pd.Series(y_test).tolist()
    

    
#     setting x range for visualisation
    x_range = np.arange(2)
    
#     creating two different array which contains spam and ham of y_pred and y_test
    y1 = [y_pred.count('spam') , y_pred.count('ham')]
    y2 = [y_test.count('spam') , y_test.count('ham')]

#     plotting bar graph
    width = 0.2
    plt.bar(x_range-0.1 , y1 , width , color = 'red')
    plt.bar(x_range+0.1 , y2 , width , color = 'green')
    
#     ticks the x axis
    plt.xticks(x_range , ["Spam" , "Ham"])
    
#     adding text in the graph
    plt.text(1.5 , -0.1 , "Accuracy Score is : \n{}".format(accuracy_score(y_pred,y_test)) , fontsize = 22)
    
#     setting labels
    plt.xlabel("Types of Messages").set_fontsize(15)
    plt.ylabel("Count of Messages").set_fontsize(15)
    plt.title("Graph of message count").set_fontsize(20)
    plt.legend(['Predicted' , 'Tested'] , bbox_to_anchor = (1.3,1))
    plt.show()
    

 




    
def Visualise_Dataset_Using_Multinomial_Algorithm(data):
    
    #     creating x and y variable of our data set column 1 contains messages and column 0 contains types of messages
    x = data["Messages"]
    y = data["Types"]
    
#     making list of different classifier to make a pipeline
    algo = [('vect' , CountVectorizer()) , ('model' , MultinomialNB())]
    
#     making pipeline
    pipe = Pipeline(algo)
    
#     spliting data and giving test size is 1/3
    x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 1/3 , random_state = 0)
    
#     fitting x_tain and y_train in created pipe
    trained = pipe.fit(x_train , y_train)
    
#     predicting the output
    y_pred = trained.predict(x_test)
    
#     converting array to list
    y_pred = np.array(y_pred).tolist()
    y_test = pd.Series(y_test).tolist()
    

    
#     setting x range for visualisation
    x_range = np.arange(2)
    
#     creating two different array which contains spam and ham of y_pred and y_test
    y1 = [y_pred.count('spam') , y_pred.count('ham')]
    y2 = [y_test.count('spam') , y_test.count('ham')]

#     plotting bar graph
    width = 0.2
    plt.bar(x_range-0.1 , y1 , width , color = 'red')
    plt.bar(x_range+0.1 , y2 , width , color = 'green')
    
#     ticks the x axis
    plt.xticks(x_range , ["Spam" , "Ham"])
    
#     adding text in the graph
    plt.text(1.5 , -0.1 , "Accuracy Score is : \n{}".format(accuracy_score(y_pred,y_test)) , fontsize = 22)
    
#     setting labels
    plt.xlabel("Types of Messages").set_fontsize(15)
    plt.ylabel("Count of Messages").set_fontsize(15)
    plt.title("Graph of message count").set_fontsize(20)
    plt.legend(['Predicted' , 'Tested'] , bbox_to_anchor = (1.3,1))
    plt.show()
    

    
    
    
    
    
def Visualise_Dataset_Using_Both_Algorithms_And_Show_Conclusion(data):
    
#     getting x and y variables from the data
    x = data["Messages"]
    y = data["Types"]
    
#     spliting data into train and test 
    x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 1/3 , random_state = 0)
    
#     creating pipeline1
    pipe1 = Pipeline([('vect' , CountVectorizer()) ,
                     ('model' , DecisionTreeClassifier())])
    
#     creating pipeline2
    pipe2 = Pipeline([('vect' , CountVectorizer()) ,
                     ('model' , MultinomialNB())])
    
#     fiting data to both pipes to tain our model
    trained1 = pipe1.fit(x_train , y_train)
    trained2 = pipe2.fit(x_train , y_train)
    
#     predicting output from trained model
    y_pred1 = trained1.predict(x_test)
    y_pred2 = trained2.predict(x_test)
    
#     calculating accuracy_score
    accu1 = accuracy_score(y_pred1,y_test)
    accu2 = accuracy_score(y_pred2,y_test)
    
#     plotting bar graph
    plt.bar(["DecisionTreeClassifier"] , [accu1] , width = 0.2)
    plt.bar(['Multinomial'] , [accu2] , width = 0.2)
    
#     setting labels and title
    plt.xlabel("Algorithms").set_fontsize(15)
    plt.ylabel("Accuracy").set_fontsize(15)
    plt.title("Comparing Accuracy Using Different Algorithms").set_fontsize(20)
    
#     adding legends
    plt.legend([accu1,accu2] , bbox_to_anchor = (1.05,1))
    
#     adding text according to comparision
    if accu1 > accu2:
        plt.text(1.5 , -0.1 , "DecisionTreeClassifier Algorithm \nis best for spam detection" , fontsize = 22)
        
    else:
        plt.text(1.5 , -0.1 , "Multinomial Naive Bayes Algorithm \nis best for spam detection" , fontsize = 22)
    
#     showing the graph
    plt.show()
    

    
    
    
# this fuction exits from the program
def Exit(data):
    sys.exit()



    
# this is main function
def main():
    
#     handiling exceptions
    try:

    #     reading dataset and storing it for further use
    #     we pass header as none because there is no columns present in dataset
        data = pd.read_csv("spam.csv" , names = ['Types','Messages'])
        
#         data messages contains various punctions marks which is not good for analysis so i run loop to modify the data
        for i in range(data['Messages'].count()):
            data.iloc[i,1] = re.sub(r'[^\w\s]','',data.iloc[i,1]).lower().strip()
            
#         initializing run variable as true it helps to run the program repeatdly according to the user
        run = True
        
#         running while loop for continous running according to the user
        while run:
    
        

        #     storing various functions in a dictionary 
            functions = {1 : Show_Dataset_Head_And_Tails,
                2 : Visualise_Current_Dataset ,
                         3 : Multinomial_Algorithm_Without_InBuilt_functions,
                        4 : Visualise_Dataset_Using_DecisonTreeClassifier_Algorithm , 
                        5 : Visualise_Dataset_Using_Multinomial_Algorithm ,
                        6 : Visualise_Dataset_Using_Both_Algorithms_And_Show_Conclusion ,
                        7 : Exit}
            
#             using for loop printing different options for the user
            for i , j in functions.items():
        
#                 here i is key and j is value of dictionary
                print(str(i)+". "+" ".join(j.__name__.split("_")))
    
#             asking from the user to choose the option from the above list
            choose = int(input("Enter which option you want to execute : "))
            
#             checking if choosen input in the functions then program execute otherwise it runs again
            if choose in functions:
                functions[choose](data)
            
            else:
                print("Choose correct option !")
                return main()
            
#             asking from the user whether he wants to run the program again or not
            is_again = input("Enter Y/y for YES else press any other key to exit : ")
            
            if is_again == 'Y' or is_again == 'y':
                continue
            
#             if user press any other then it comes to this else condition and set run as False which breaks the loop
            else:
                print("Exitted")
                run = False
                break
                
#     this exception occurs if user press stop button 
    except KeyboardInterrupt:
        print("Program Stopped")

#         this exception ocuurs if user exits from the program
    except SystemExit:
        print("Exitted from the program")

#         this exception arise this means that use not entered number in choose
    except:
        print("Choose a number")
        return main()


# calling main functin
if __name__ == "__main__":
    main()
```
